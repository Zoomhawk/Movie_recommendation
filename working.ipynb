```python
# genreObj = movies['genres'][0]
# print( type(genreObj) )
import json

def getGenres(genrStr):
    temp = json.loads(genrStr)
    ans = []
    for obj in temp:
        ans.append(obj['name'])
    return ans

def getTopPeople(genrStr):
    temp = json.loads(genrStr)
    ans = []
    for obj in temp:
        ans.append(obj['name'])
    return ans[:5]
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
ps = PorterStemmer()

def getStemWordList(words):
    result = []
    for word in words :
        result.append(ps.stem(word))
    return result
```

```python
new_df['movieTags'] = new_df['movieTags'].map(getStemWordList)
```

```python
large_list = []
for tags in new_df['movieTags']:
    large_list.extend(tags)
```

```python
from collections import Counter

word_freq = Counter(large_list)

temp = word_freq.most_common(5001)
dimenstionArray = []
for tup in temp:
    dimenstionArray.append(tup[0])
dimenstionArray.pop(2) # pta nhi kyu iss position pe "" aa rha tha toh uda diya
print(len(dimenstionArray))
```

```python
def vectorize(arr):
    freq = Counter(arr)
    vector = []
    for dimention in dimenstionArray:
        if(dimention not in freq):
            vector.append(0)
        else:
            vector.append(freq[dimention])
    return vector
```

```python
new_df['vector'] = new_df['movieTags'].map(vectorize)
new_df.drop('movieTags', axis=1, inplace=True)
new_df.head()
```

```python
from math import sqrt

def cosine_similarity(l1, l2):
    dot_product = sum(i*j for i,j in zip(l1, l2))
    magnitude1 = sqrt(sum(i**2 for i in l1))
    magnitude2 = sqrt(sum(j**2 for j in l2))
    if(magnitude1 == 0):
        magnitude1 = 1
    if(magnitude2 == 0):
        magnitude2 = 1
    return dot_product / (magnitude1 * magnitude2)
```

```python
def top3Movies(movieID):
    movieVector = getMovieVector(movieID)
    simmilarityArray = []
    simMovID = []
    for vec in new_df['vector']:
        simmilarityArray.append(cosine_similarity(movieVector,vec))
    for id in new_df['movieId']:
        simMovID.append(id)
    
    top1score = 0
    top2score = 0
    top3score = 0
    top1 = -1
    top2 = -1
    top3 = -1

    for i in range(4799):
        if(simMovID[i] == movieID):
            continue

        if(simmilarityArray[i] > top1score):
            top3score = top2score
            top3 = top2
            top2score = top1score
            top2 = top1
            top1score = simmilarityArray[i]
            top1 = simMovID[i]
        elif(simmilarityArray[i] > top2score):
            top3score = top2score
            top3 = top2
            top2score = simmilarityArray[i]
            top2 = simMovID[i]
        elif(simmilarityArray[i] > top3score):
            top3score = simmilarityArray[i]
            top3 = simMovID[i]

    return [top1,top2,top3]


```
